```python
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot as plt
import seaborn as sns
```

## Xb = (X1 + X2 + ... + Xn) / n を作る

```python
n = 10   # 互いに独立なn個の確率変数X1, X2, X3, ..., Xnの個数
N = 500  # Xnのサイズ

X = np.random.normal(loc=30, scale=500, size=(n, N))
```

```python
Xb = np.mean(X, axis=0)  # 列ごとの平均
```

```python
# mu = μ  
print('mu: {}'.format(np.mean(Xb)))

# sigma2 = σ^2/n
print('sigma2: {}'.format(np.var(Xb)))
```

## 正規性の確認

```python
# Q-Qプロット
stats.probplot(Xb, dist="norm", plot=plt)
plt.show()
```

```python
sns.distplot(Xb)
```
