# 演習問題7 独立な確率変数の期待値と分散(I)

```python
import numpy as np
from scipy.stats import norm
import pandas as pd
import scipy.stats as stats
import math

from matplotlib import pyplot as plt
import seaborn as sns; sns.set() # sns.set() ==> グラフの見た目をseabornに合わせる
```

## (1) Xの期待値と分散

```python
N = 1000  # シミュレーションのサンプルサイズ

# 正規分布と仮定
T = np.random.normal(loc=1, scale=math.sqrt(2), size=N)
U = np.random.normal(loc=-1, scale=math.sqrt(1), size=N)
```

```python
sns.distplot(T);
sns.distplot(U);
```

```python
X = 2*T + U + 2
sns.distplot(T);
sns.distplot(U);
sns.distplot(X);
```

```python
# Xの標本平均、標本分散、ヒストグラム
meanX = np.mean(X)  # 2*E[T] + E[U] + 2 = 3
varX = np.var(X)    # 2**2*V[T] + V[U] = 9
print('mean={}, var={}'.format(meanX, varX))
```

## (2) bar(X)の期待値と分散

```python
n = 100
Xn = np.random.normal(loc=meanX, scale=math.sqrt(varX), size=(n,N))
```

```python
Xn.shape
```

```python
barX = np.mean(Xn, axis=0)
barX.shape
```

```python
# Xの標本平均、標本分散、ヒストグラム
meanbarX = np.mean(barX)
varbarX = np.var(barX)
print('mean={}, var={}'.format(meanbarX, varbarX))
```

varが9/nになってる。
