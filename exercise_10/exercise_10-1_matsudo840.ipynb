```python
import numpy as np
from scipy.stats import poisson
import random

from matplotlib import pyplot as plt
import seaborn as sns; sns.set() # sns.set() ==> グラフの見た目をseabornに合わせる
```

平均μ=3のポアソン分布Po(3)に従う乱数を発生させて、P(X≦5)の確率をシミュレーションで求める。

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html
# https://qiita.com/g-k/items/836820b826775feb5628

x = np.arange(15)
P = [poisson.pmf(i, 3) for i in x]
sns.barplot(x, P)
```

```python
N = 100000   # 試行回数

result = [0]*15
for i in range(N):
    rand_x = random.randint(0, 14)
    rand_y = random.random()
    if rand_y <= P[rand_x]:
        result[rand_x] += 1

sns.barplot(x, result)
```

```python
print(sum(result[5:])/sum(result))
```

P(X≦5) = 0.1847
