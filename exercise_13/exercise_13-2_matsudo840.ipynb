```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## X2分布を作る

```python
n = 10   # 互いに独立なn個の確率変数Z1, Z2, Z3, ..ニジョウしてから
N = 200  # Znのサイズ

Z = np.random.normal(size=(n, N))
```

```python
Y = Z**2  # 要素を二乗
X = np.cumsum(Y, axis=0)  # 列ごとに累積和
```

```python
for i in range(n):
    print('n: {}, Ec[X]: {:.3f}, Vc[X]: {:.3f}'.format(i+1, np.mean(X[i, :]), np.var(X[i, :])))
```

```python
X1 = X[0]
X2 = X[1]
X3 = X[2]
X5 = X[4]
X10 = X[9]

# Plot
# kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})

plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(X1, color="dodgerblue", label="1", **kwargs)
sns.distplot(X2, color="orange", label="2", **kwargs)
sns.distplot(X3, color="deeppink", label="3", **kwargs)
sns.distplot(X5, color="green", label="5", **kwargs)
sns.distplot(X10, color="purple", label="10", **kwargs)

plt.xlim(-5,20)
plt.legend();
```
