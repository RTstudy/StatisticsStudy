# 演習問題5 連続型確率変数の期待値と分散(I)

```python
import numpy as np
import math
from scipy.stats import norm
import pandas as pd
import scipy.stats as stats

from matplotlib import pyplot as plt
import seaborn as sns; sns.set() # sns.set() ==> グラフの見た目をseabornに合わせる
```

```python
start = 0
end = math.pi
n = 1000  # 分割数
X = np.linspace(start, end, n)
```

```python
sns.distplot(X);
```

```python
def f(x):
    return math.sin(x)/2
```

```python
fx = list(map(f, X))
sns.distplot(fx);
```

```python
## 失敗コード
# meanY = np.mean(Y)
# varY = np.var(Y)
# print('mean={}'.format(meanY))
# print('var={}, correct answer={}'.format(varY, math.pi**2/4-2))
```

```python
Y = X*fx
meanY = np.mean(Y)
varY = np.var(Y)
print('mean={}'.format(meanY))
print('var={}, correct answer={}'.format(varY, math.pi**2/4-2))
```
