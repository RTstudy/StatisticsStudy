# 演習問題5 連続型確率変数の期待値と分散(I)

```python
import numpy as np
import math
from scipy.stats import norm
import pandas as pd
import scipy.stats as stats

from matplotlib import pyplot as plt
import seaborn as sns; sns.set() # sns.set() ==> グラフの見た目をseabornに合わせる
```

## 普通に解く

```python
start = 0
end = math.pi
n = 1000  # 分割数
df = pd.DataFrame()
df['X'] = np.linspace(start, end, n)
```

```python
sns.distplot(df['X']);
```

```python
def f(x):
    if 0<=x<=math.pi:
        return math.sin(x)/2
    else:
        return 0
```

```python
df['fX'] = df['X'].map(f)
sns.relplot(x='X', y='fX', kind='line', data=df);
```

```python
df['Y'] = df['X'] * df['fX'] * math.pi / n
```

```python
myu = sum(df['Y'])
myu
```

## von Neumannの棄却法で解く

see. issue #5

```python
n = 10000
Z = np.random.rand(n, 2)

X = []
for z in Z:
    if z[1] < np.sin(z[0]*np.pi)/2:
        X.append(z[0]*np.pi) 
```

```python
sns.distplot(X)
```
